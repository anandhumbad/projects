```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
```

```python
df=pd.read_excel('Data Visualization Assignmnet.xlsx')
```

```python
df.head(20)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
sns.boxplot(data=df['UNIT'],orient='h')
plt.title('No of units GRAPH')
plt.xlabel('VALUE')
plt.ylabel('PLOT')
```

THE DATASET IS POSITIVLY SKEWED .

```python
x=df['DESCRIPTION']
y=df['UNIT']
plt.figure(figsize=(10,6))
plt.scatter(x,y, c='red')
plt.title('scatter plot of NO. of units VS Products ')
plt.ylabel(' No of Units')
plt.xlabel(' PRODUCTS')
```

```python
features =  ['DESCRIPTION']

for col in features:
  sns.displot(df, x=col, kde=True)
plt.figure(figsize=(40,40))
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title("No of Units ON a specific Date", fontsize=15)
sns.lineplot(x='DATE', y='UNIT', data=df)
# CAN be used to determine on which dates most products are ordered.
```

```python
plt.rcParams["figure.figsize"] = (20,20)
sns.barplot(y =df['PLACE'],x =df['UNIT'], palette='rocket_r')
plt.title('ORDERS NO BY CITES')
plt.show()
#TOTAL no of orders by cities.
```

```python
plt.figure(figsize=(20,20))
plt.title("Quantity of products ON a specific Date", fontsize=15)
sns.lineplot(x='DESCRIPTION', y='UNIT', data=df)
plt.xlabel('PRODUCTS')
```

```python
sns.countplot(data=df, x='UNIT', hue='PLACE', palette = 'Set3')
plt.rcParams["figure.figsize"] = (2,2)
plt.show()
```

```python
plt.figure(figsize=(17,17))
sns.lineplot(x='DESCRIPTION', y='UNIT', hue ='PLACE', data=df)
plt.xlabel('PRODUCTS')
```

```python
sns.barplot(data=df, x='UNIT', y='DATE', palette='coolwarm')
plt.rcParams["figure.figsize"] = (20,20)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar("DESCRIPTION", "UNIT", data = df, color = "blue")
plt.xlabel("PRODUCTS")
plt.xticks(rotation = 90)
plt.ylabel("UNITS")
plt.title("NO of Units of produts Ordered")
plt.show()
```
